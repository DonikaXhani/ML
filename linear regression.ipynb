```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x=np.array([1,10,20,40,50,70,80,90,120])
y=np.array([3,20,90,110,130,170,150,200,260])
```

```python
lr=LinearRegression()
```

```python
#preprocess the data
x=x.reshape(-1,1)
```

```python
lr.fit(x,y)
```

```python
y_pred=lr.predict(x)
```

```python
#make a scatterplot
plt.scatter(x,y)
#fit the linear regression line on the scatterplot
plt.plot(x,y_pred, color='red')
plt.show()
```

```python
#find the equation of the linear regression
print(lr.coef_)
print(lr.intercept_)
```

```python
#multiple linear regression
import pandas as pd
#load the data
d = pd.read_csv('Folds5x2_pp.csv')
```

```python
d.head()
```

```python
#define x and y
x=d.drop(['PE'],axis=1).values
y=d['PE'].values
```

```python
print(x)
```

```python
print(y)
```

```python
#split the dataset in training and testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=0)
```

```python
#train the model on the training set
ml=LinearRegression()
ml.fit(x_train,y_train)
```

```python
#predict the test results
y_pred=ml.predict(x_test)
print(y_pred)
```

```python
#evaluate the model through R^2
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
#plot the results
plt.figure(figsize=(15,10))
plt.scatter(y_test, y_pred)
plt.xlabel('actual')
plt.ylabel('predicted')
plt.title('Actual vs predicted')
```
